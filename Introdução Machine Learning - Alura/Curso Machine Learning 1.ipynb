Primeiros testes (Comparando Porcos e Cachorros)

```python
#caracteristicas "features" sao essenciais para o algoritmo aprender melhor 
#pelo longo?
#perna curta?
#faz auau?

porco1 = [0,1,0]
porco2 = [0,1,1]
porco3 = [1,1,0]

cachorro1 = [0,1,1]
cachorro2 = [1,0,1]
cachorro3 = [1,1,1]

treino_x = [porco1,porco2,porco3,cachorro1,cachorro2,cachorro3]
# 1 => Porco, 0=> cachorro
treino_y = [1,1,1,0,0,0] #label
```

```python
from sklearn.svm import LinearSVC
#usando o LinearSVC o modelo vai tentar se encaixar com base nos dados e classes
model = LinearSVC()
model.fit(treino_x,treino_y)
```

```python
#Primeiro teste de predição de um animal 
animal_misterioso = [1,1,1]
#Predict sempre espera um Array 
model.predict([animal_misterioso])
```

```python
mist1 = [1,1,1]
mist2 = [0,1,0]
mist3 = [0,1,1]
#prevendo a classe de 3 animais 
teste_x = [mist1,mist2,mist3]
teste_y = [0,1,1]
previsoes = model.predict(teste_x)
```

```python
#calculando a taxa de acerto do algoritmo 
corretos = (previsoes == teste_y).sum()
total = len(teste_x)
acertos = corretos/total
print("Taxa de acertos %.2f " %(acertos *100))
```

```python
#taxa de acerto atraves do accuracy_score do sklearn 
from sklearn.metrics import accuracy_score

acerto = accuracy_score(teste_y,previsoes)
print("Taxa de acerto %.2f"%( acerto * 100))
```
